```
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```
df_decompA = pd.read_csv("/content/distances_decomp_A.csv")
df_decompB = pd.read_csv("/content/distances_decomp_B.csv")
df_decompC = pd.read_csv("/content/distances_decomp_C.csv")
```

# **Cumulated distances** - Bar Chart

```
def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width()/2., 0.5*height,
                '%.3f' % float(height),
                ha='center', va='bottom')
```

```
decompA_sum = df_decompA["distance"].sum()
decompB_sum = df_decompB["distance"].sum()
decompC_sum = df_decompC["distance"].sum()

labels = ['decompA', 'decompB', 'decompC']
distances = [decompA_sum, decompB_sum, decompC_sum]
distances_bar_plt = plt.bar(labels, distances, align='center', alpha=0.5)
plt.rcParams["figure.figsize"] = (3,5)
plt.ylabel('Cumulated distance')
plt.xlabel("Decompression algorithms")
plt.title('Decompression algorithms total distances')
autolabel(distances_bar_plt)
plt.plot()
```

#**Standard deviation** - Boxplot

```
decompA_std = df_decompA['distance'].std()
decompB_std = df_decompB['distance'].std()
decompC_std = df_decompC['distance'].std()
labels = ['decompA', 'decompB', 'decompC']
distances = [df_decompA['distance'], df_decompB['distance'], df_decompC['distance']]

plt.rcParams["figure.figsize"] = (15,5)
fig, (ax1, ax2, ax3) = plt.subplots(1,3)
ax1.boxplot(df_decompA['distance'], notch=False)
ax2.boxplot(df_decompB['distance'], notch=False)
ax3.boxplot(df_decompC['distance'], notch=False)
ax1.set_title('Decomp Algorithm A Box')
ax2.set_title('Decomp Algorithm B Box')
ax3.set_title('Decomp Algorithm C Box')
plt.show()
```
